All great circles are geodesics. Geodesics are curves that locally minimizes arc-length. Given any two points on a great circle, the great circle locally minimizes the distance between the two points and "shifting" that great circle in any manner will increase the distance between the two points.


```
# preprocessing
import numpy as np
np.random.seed(0)
# object = 2xn   # 2 because x,y and n is the number of points of shape

###
## Preshape
###

# Remove Translation: Subtract row mean from each row
def remove_translation(obj):
  return obj - obj.mean(axis=1, keepdims=True)


# Remove Scale: Divide matrix by Frobenius Norm
def remove_scale(obj):
  frob_norm = np.linalg.norm(obj, "fro")
  obj = obj/frob_norm
  return obj

def get_preshape(obj):
    obj = remove_translation(obj)
    obj = remove_scale(obj)
    return obj
```

```
# Orthogonal Procustes Analyses - remove rotation
def opa(target_preshape, moving_preshape):
    combined_preshape = np.matmul(target_preshape, moving_preshape.transpose())
    U, sigma, V = np.linalg.svd(combined_preshape)
    R_star = np.matmul(U,V.T)

    # If det(R_Star) is negative, then swap columns...
    # technically wrong, but ok for this assignment
    # Use instead a Constrained OPA with kabach algo
    if np.linalg.det(R_star)<0:
        R_star[:,[0, 1]] = R_star[:,[1, 0]]
    # print("R_STAR: ", R_star)
    return R_star
```

```
import math

def exp_map(preshape, tangent_vector):
    # If tangent vector is zero vecttor
    if not np.any(tangent_vector):
        return preshape
        
    v = tangent_vector
    v_norm = theta = np.linalg.norm(tangent_vector, ord=2)
    q = preshape*math.cos(theta) + (v/v_norm) * math.sin(theta)
    return q

def log_map(p,q):
    # Do OPA Before Log Map
    rot = opa(p,q)
    q = rot @q
    # Floating point inaccuracy can lead to values above 1
    dot_prod = min(np.dot( p.flatten(), q.flatten()), 1)
    v_norm = theta = np.arccos(dot_prod)

    # We can use Gram Schmidt to do this without worrying about 0 v_norm
    if v_norm == 0:
        return (q- p * math.cos(theta)) * v_norm

    v = (q- p * math.cos(theta))/math.sin(theta) * v_norm
    return v
```

```
# frechetet_mean(intrinsic mean)
# STOP Via epoch or when SIGMA is small
def frechet_mean(manifold_points, epoch):
  mu = manifold_points[0]
  for i in range(epoch):
    sigma_log_map = np.zeros(manifold_points[0].shape)
    for x_i in manifold_points:
      logmap = log_map(mu,x_i)
      sigma_log_map += logmap
    
    sigma_mu = (1/len(manifold_points)) * sigma_log_map
    mu = exp_map(mu, sigma_mu)
  return mu
```

```
def PGA(X, fre_mean):
    '''
    :X: nxd data matrix
    '''
    n = len(X)
    log_map_sum = []
    for i in range(len(X)):
        u_i_log_map = log_map(fre_mean, X[i])
        log_map_sum.append(u_i_log_map.flatten())
    X = np.array(log_map_sum)

    covariance_matrix = (1/n) * np.matmul(X.T, X)
    print(covariance_matrix.shape)
    # eighenvalue, eigenvectors
    
    w, v = np.linalg.eigh(covariance_matrix)
    print(w.shape, v.shape)
    return w, v
```

```
# 1. Generate Random Triangles
import numpy as np
def gen_rand_triangle(n):

    triangle_list = np.empty((0,2,3))
    for i in range(n):
        p_0 = [-1,0]
        p_1 = [1,0]

        # mu = mean, sigma = std deviation
        mu, sigma = 1, 0.5
        s = np.random.normal(mu, sigma)
        p_2 = [0,s]


        # Add noise
        # Same noise to all or no?
        for point in [p_0, p_1, p_2]:
            noise1 = np.random.normal(0, 0.1)
            noise2 = np.random.normal(0, 0.1)

            point[0] += noise1
            point[1] += noise2
        
        # Appending list to ndarray converts list to ndarray automatically, but we want to transpose matrix, which is easy for ndarray
        triangle = np.asarray([p_0,p_1,p_2]).T
        triangle_list = np.append(triangle_list, [triangle], axis=0)
    return triangle_list
```
