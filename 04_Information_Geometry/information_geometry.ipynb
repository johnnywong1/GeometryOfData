```
def geo_dist_2(p1, p2):
    x1, y1 = p1
    x2, y2 = p2

    b = (x1 + x2)/2
    a = math.sqrt(2 * y1**2 + (x2 - b)**2)
    c = np.arctanh((x2 - b)/a)

    # v = np.array([a*c*(1/np.cosh(c * t))**2, -(a * c * (1/np.cosh(c*t)) * np.tanh(c*t))/np.sqrt(2) ])
    # g = np.array([ [1/(-(a*c *(1/np.cosh(c*t))*np.tanh(c*t))/np.sqrt(2))**2, 0] , [0, 2/(-(a*c *(1/np.cosh(c*t))*np.tanh(c*t))/np.sqrt(2))**2]])

                    
    g_dist = lambda t: np.sqrt(np.array([a*c*(1/np.cosh(c*t))**2, -(a * c * (1/np.cosh(c*t)) * np.tanh(c*t))/np.sqrt(2) ]) @ np.array([ [1/(a*(np.sqrt(2)/2)*(1/np.cosh(c*t)))**2, 0] , [0, 2/(a*(np.sqrt(2)/2)*(1/np.cosh(c*t)))**2]]) @ np.array([a*c*(1/np.cosh(c*t))**2, -(a * c * (1/np.cosh(c*t)) * np.tanh(c*t))/np.sqrt(2)]).T)

    t0, t1 = -1, 1
    return integrate.quad(g_dist, t0, t1)

geo_dist_2((-1,1), (1,1))
print('Geodesic Distance:', (g_dist[0]))
```

```
def find_midpoint(p1, p2):
    # Note this midpoint is also the Frechet mean between the two gaussian
    x1, y1 = p1
    x2, y2 = p2

    # Case 1: problem 1(d) whose geodesic curve is 1(b)
    if x1 == x2: 
        a = x1  # x1 = x2 = 0 in 1(d)  
        b = y1  
        c = math.log(y2) - math.log(b) 
        t = 0.5
        return [a, b * math.exp(c * t)]
    else:
        b = (x1 + x2)/2
        a = math.sqrt(2 * y1**2 + (x2 - b)**2)
        c = np.arctanh((x2 - b)/a)
        t = 0
        return [a * np.tanh(c*t), a*(math.sqrt(2)/2)*(1/np.cosh(c*t))]
```

```
p1_e1, p2_e1 = (-1,1),(1,1)
midpoint_1e = find_midpoint(p1_e1, p2_e1)

gaus_list = []
# Mean = -1, SD = 1
gaus_list.append((-1,1))
# Mean = 1, SD = 1
gaus_list.append((1,1))
#midpoint (mu, sigma)
gaus_list.append((midpoint_1e[0],midpoint_1e[1]))

plot_pdf(gaus_list)
plot_poincare((-1,1), (1,1), midpoint=True)
```

The resulting gaussian created from the frechet mean of the two gaussian has a mean at 0 or the center between the two gaussians means.

```
torch.manual_seed(0)
class LogisticRegression(torch.nn.Module):
    def __init__(self):
        super(LogisticRegression, self).__init__()
        self.linear = torch.nn.Linear(784, 1, bias=False)     
        torch.nn.init.normal_(self.linear.weight, mean=0, std=1.0)

    def forward(self, x):
        x = x.view(-1, 784)
        y_pred = torch.sigmoid(self.linear(x))
        return y_pred
        
model = LogisticRegression()

criterion = torch.nn.BCELoss() #CrossEntropyLoss()
```

```
import matplotlib.pyplot as plt
import copy

# Baseline to compare with
pytorch_copy_model = copy.deepcopy(model)
# Logistic Regression
def standard_gradient_logreg(X_train, y_train, model, lr, iterations):
    optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)
    
    model.train()
    loss_list = []
    for epoch in range(iterations):
        model.linear.zero_grad()
        y_pred = model(X_train.float())    # Compute Loss

        # Update weights with gradient
        loss = criterion(y_pred.flatten(), y_train.float())  # Cross-Entropy Loss  
        # print(loss)
        loss_list.append(loss)
        loss.backward() # equivalent to np.dot(X.T, y_pred - labels)
        optimizer.step()
    plt.plot(loss_list)
    plt.show()
    print(loss_list[-1])
standard_gradient_logreg(X_train, y_train, pytorch_copy_model, 0.1, 20)
```

```
import matplotlib.pyplot as plt

# Logistic Regression
standard_copy_model = copy.deepcopy(model)

def standard_gradient_logreg(X_train, y_train, model, lr, iterations):
    model.train()
    loss_list = []
    for epoch in range(iterations):
        model.linear.zero_grad()
        y_pred = model(X_train.float())    # Compute Loss

        # Update weights with gradient
        loss = criterion(y_pred.flatten(), y_train.float())  # Cross-Entropy Loss  
        # print(loss)
        loss_list.append(loss)
        loss.backward() # equivalent to np.dot(X.T, y_pred - labels)
        weight_gradient = model.linear.weight.grad
        with torch.no_grad():
            model.linear.weight -= lr * weight_gradient
    plt.plot(loss_list)
    plt.show()
    print(loss_list[-1])


standard_gradient_logreg(X_train, y_train, standard_copy_model, 0.1, 20)
```

```
import matplotlib.pyplot as plt

# Natural Gradient
natural_copy_model = copy.deepcopy(model)
def natural_gradient_logreg(X_train, y_train, model, lr, iterations):
    model.train()
    loss_list = []
    for epoch in range(20):

        # We are doing batch gradient descent
        fisher_matrix = torch.zeros(784, 784)
        average_gradient = torch.zeros(1, 784)
        average_loss = 0
        for i in range(X_train.shape[0]):
            model.linear.zero_grad()
            y_pred = model(X_train.float()) 
          
            loss = criterion(y_pred.flatten(), y_train.float())  # Cross-Entropy Loss  
            average_loss += loss

            loss.backward() # This gives you standard gradient
            gradient = model.linear.weight.grad
            average_gradient += gradient

            grad_square = (gradient @ gradient.T)
            fisher_matrix += grad_square
        fisher_matrix /= X_train.shape[0]
        average_gradient /= X_train.shape[0]
        loss_list.append(average_loss/X_train.shape[0])
        
        fisher_matrix_inv = torch.pinverse(fisher_matrix)
        natural_gradient = torch.matmul(fisher_matrix_inv, average_gradient.T)

        with torch.no_grad():
            model.linear.weight -= lr * natural_gradient.T
    plt.plot(loss_list)
    plt.show()

natural_gradient_logreg(X_train, y_train, natural_copy_model, 0.2, 20)
```
