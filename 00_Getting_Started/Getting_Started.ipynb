Given a metric space $X$ with metric $d: X * X \rightarrow \mathbb{R}$, its basis will be $\mathbb{B_X}$ which is composed of open balls in the format $Br(x,r)=${$y \in X: d(x,y) < r$} for all $x \in X, r \in \mathbb{R}$.

$\textbf{Prove:} $ that the collection of open sets {$B \cap A: B \in \mathbb{B_X}$} form a basis, $\mathbb{B_A}$, for the subspace toplogy on A.

Because we are given that a subset $A \subseteq X$, we can say that $\forall a \in A$, $a \in X$. Because $a \in X$ and $\mathbb{B_X}$ is a basis for our metric space $X$, we cay say $\exists B \in \mathbb{B_x}$ such that $a \in B$. Thus $\forall a \in A, \exists (B \cap A)$ such that $a \in (B \cap A)$ and since $(B \cap A)$ are elements of $\mathbb{B_A}, $this satisfies the first requirement of a basis

Assume we have arbitrary elements $B_1,B_2$ in $\mathbb{B_A}$ and an $x \in X$ such that $x \in B_1 \cap B_2$. From our definition of $\mathbb{B_A}$, we can state that  $B_1$ and $B_2$ are elements equal to $(B^{1}_1 \cap A)$ and $(B^{1}_2 \cap A)$ respectively with $B^{1}_1, B^{1}_2 \in \mathbb{B_X}$. With this knowledge, we can do the following: 

\begin{align}
    a &\in (B_1 \cap B_2)\\
    a &\in (B^{1}_1 \cap A) \cap (B^{1}_2\cap A)\\
    a &\in (B^{1}_1 \cap B^{1}_2) \cap A \\
    a &\in (B_3) \cap A \\
    a &\in (B_3)\\
\end{align}
We can conclude that $B_3 \subset (B^{1}_1 \cap B^{1}_2)$ from step 3 to 4 as we know $\mathbb{B_X}
$ is a basis, $B^{1}_1, B^{1}_2 \in \mathbb{B_X}$, and $x \in B^{1}_1 \cap B^{1}_2$..


Thus, if $B_1, B_2 \in \mathbb{B_A}$ and $x \in B_1 \cap B_2$, then there indeed exists a $B_3 \subset B_1 \cap B_2$ such that $x \in B_3$. 

We can conclude that indeed the collection of open sets {$B \cap A: B \in \mathbb{B_X}$} form a basis, $\mathbb{B_A}$, for the subspace toplogy on A.

Given this information, with the fact that the $B$s in $\mathbb{B_X}$ are in the format $Br(x,r)=${$y \in X: d(x,y) < r$} for all $x \in X, r \in \mathbb{R}$. and the fact that only those that intersect with $A$ are in $\mathbb{B_A}$, we can say that the open elements in $\mathbb{B_A}$ are in the format $Br(x,r)=${$y \in A: d(x,y) < r$} for all $x \in X, r \in \mathbb{R}$. This is conveniently what the basis of the metric space X restricted A is also composed of, meaning the topologies created from both are the same.

```
import numpy as np
def sample_d_gaussian_randvar(d):
  for i in range(0, d+1):
    yield np.random.normal(0,1)

# Test above
# print(*sample_d_gaussian_randvar(2))
```

```
def project_points(d, num_pts, equator = False):
  point_list = []
  equator_list = []
  for i in range(num_pts):
    x = list(sample_d_gaussian_randvar(d))
    equator_x = x.copy()
    equator_x[0] = 0

    x_norm = np.linalg.norm(x)
    equator_norm = np.linalg.norm(equator_x)

    unit_sphere_x = [x_i/x_norm for x_i in x]
    unit_sphere_equator = [x_i/x_norm for x_i in equator_x]
    point_list.append(unit_sphere_x)
    equator_list.append(unit_sphere_equator)
  return point_list, equator_list
```

```
import math
import matplotlib.pyplot as plt

d_list = [2, 4, 10, 100, 1000]
for d in d_list:
  proj_points, proj_equator = project_points(d, 10000)
  # euclid_dist = [np.linalg.norm(np.array(p_p) - np.array(p_j)) for p_p, p_j in zip(proj_points, proj_equator)]
  
  # radius = 0.5
  # central_angle = [ 2 * math.asin(e_dist/2*radius) for e_dist in euclid_dist]

  # great_circle_dist = [0.5 * c_angle for c_angle in central_angle]

  great_circle_dist = [ math.acos(np.dot(np.array(p_p),np.array(p_j))) for p_p, p_j in zip(proj_points, proj_equator) ]
  print(great_circle_dist)
  plt.hist(great_circle_dist, bins='auto')
  plt.title(d)
  plt.show()
# Central Angle:
# https://math.stackexchange.com/questions/3318680/finding-angle-between-two-points-on-circle-given-cartesian-coordinates-of-the-po
```

As d increases, the graph screws more to the left. This suggests that the uniform random points on the sphere approach distance 0 from the equator as goes to infinity. From some test cases, I find that the result will be same even if we used a different great circle.
